```python
import osmnx as ox
import networkx as nx
from IPython.display import IFrame

from geopy.geocoders import Nominatim #To gather the coordinates
from pprint import pprint

ox.config(log_console=True, use_cache=True)
```

```python
location_name_1 = "Luyego"
location_name_2 = "Tabuyo del Monte"

geolocator = Nominatim(user_agent="geoapiExercises_IFrame")
```

```python

Place_1 = location_name_1
Place_2 = location_name_2

print (geolocator.geocode(Place_1))
location_1 = geolocator.geocode(Place_1)
location_2 = geolocator.geocode(Place_2)

print (location_1.longitude)

long_1 = ("Longitude: ", location_1.longitude)
Lat_1 = ("Latitude: ", location_1.latitude)
long_2 = ("Longitude: ", location_2.longitude)
Lat_2 = ("Latitude: ", location_2.latitude)

#Assigning new variable names
coord_1 = [location_1.latitude, location_1.longitude]
coord_2 = [location_2.latitude, location_2.longitude]
```

```python


G_walk = ox.graph_from_place('Luyego', network_type='all_private')
#  G_walk = ox.graph_from_place("Tabuyo del Monte, Luyego, León, Castilla y León", network_type="all_private")
# fig, ax = ox.plot_graph(G)
# ox.distance.nearest_nodes

# plot the street network with folium
m1 = ox.plot_graph_folium(G_walk, popup_attribute="name", weight=2, color="#8b0000")


print(long_1)
orig_node = ox.nearest_nodes(G_walk, location_1.longitude,location_1.latitude)

dest_node = ox.nearest_nodes(G_walk, coord_2[1], coord_2[0])
print (dest_node)
route = nx.shortest_path(G_walk, orig_node, dest_node, weight='length')
print (route)

route_map = ox.plot_route_folium(G_walk, route)

route_map
```

```python
# save as html file then display map as an iframe
filepath = "output_data/route.html"
route_map.save(filepath)
IFrame(filepath, width=600, height=500)
```
